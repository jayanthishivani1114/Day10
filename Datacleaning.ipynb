```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import zscore
```

```python
data=pd.read_csv(r"C:\Users\VIJAYAKUMARA\OneDrive\Documents\Downloads\sample2.csv")
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
fig, axs=plt.subplots(1,4,figsize=(16,6))
axs[0].boxplot(data.Age)
axs[1].boxplot(data.Salary)
axs[2].boxplot(data.Experience)
axs[3].boxplot(data.Monthly_Sales)
plt.show()
```

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit =Q1-1.5*IQR
```

```python
outliers = ((data < lower_limit ) | (data > upper_limit)).any(axis=1)
outlier_data = data[outliers]
cleaned_data = data[~outliers]
```

```python
cleaned_data.head()
```

```python
fig, axs=plt.subplots(1,4,figsize=(16,6))
axs[0].boxplot(cleaned_data.Age)
axs[1].boxplot(cleaned_data.Salary)
axs[2].boxplot(cleaned_data.Experience)
axs[3].boxplot(cleaned_data.Monthly_Sales)
plt.show()
```

```python
data.head()
```

```python
z_scores=np.abs(zscore(data))
z_score_mask=(z_scores>3).any(axis=1)
data[z_score_mask]
```
